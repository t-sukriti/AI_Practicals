```python
#Practical 1
def generate_magic_square(n):
    # Check if n is odd
    if n % 2 == 0:
        print("Magic square is only possible for odd numbers!")
        return

    # Initialize n x n matrix with 0
    magic_square = [[0 for _ in range(n)] for _ in range(n)]

    # Initial position
    i = 0
    j = n // 2

    # Fill the magic square
    for num in range(1, n*n + 1):
        magic_square[i][j] = num

        # Move up and right
        new_i = (i - 1) % n
        new_j = (j + 1) % n

        # If already filled, move down instead
        if magic_square[new_i][new_j] != 0:
            i = (i + 1) % n
        else:
            i, j = new_i, new_j

    return magic_square


# Input from user
n = int(input("Enter an odd number (size of magic square): "))

# Generate magic square
square = generate_magic_square(n)

# Display result
if square:
    print("\nMagic Square:")
    for row in square:
        for num in row:
            print(f"{num:3}", end=" ")
        print()

    # Verify magic sum
    magic_sum = n * (n*n + 1) // 2
    print("\nMagic Sum =", magic_sum)
```

```python
#Practical 2
# Water Jug Problem using Production Rules (State Space Search)

from collections import deque

# Define capacities
CAP_A = 4
CAP_B = 3

# Goal: 2 gallons in Jug A
GOAL = 2

# Function to generate all possible next states (Production Rules)
def get_next_states(a, b):
    states = []

    # Fill Jug A
    states.append((CAP_A, b))

    # Fill Jug B
    states.append((a, CAP_B))

    # Empty Jug A
    states.append((0, b))

    # Empty Jug B
    states.append((a, 0))

    # Pour A -> B
    transfer = min(a, CAP_B - b)
    states.append((a - transfer, b + transfer))

    # Pour B -> A
    transfer = min(b, CAP_A - a)
    states.append((a + transfer, b - transfer))

    return states


def solve_water_jug():
    visited = set()
    queue = deque()

    # Initial state (0,0)
    queue.append((0, 0, []))

    while queue:
        a, b, path = queue.popleft()

        # Goal check
        if a == GOAL:
            return path + [(a, b)]

        if (a, b) in visited:
            continue

        visited.add((a, b))

        # Apply production rules
        for next_state in get_next_states(a, b):
            if next_state not in visited:
                queue.append((next_state[0], next_state[1], path + [(a, b)]))

    return None


# Run solution
solution = solve_water_jug()

# Display result
if solution:
    print("Steps to reach goal:\n")
    for step in solution:
        print(f"Jug A: {step[0]} gallons, Jug B: {step[1]} gallons")
else:
    print("No solution found.")
```

```python
#Practical 3
import time
import heapq
from collections import deque

graph = {
    'A': [('B', 1), ('C', 3)],
    'B': [('D', 3), ('E', 1)],
    'C': [('F', 5)],
    'D': [],
    'E': [('F', 1)],
    'F': []
}

heuristic = {
    'A': 6,
    'B': 4,
    'C': 4,
    'D': 2,
    'E': 1,
    'F': 0
}

start = 'A'
goal = 'F'

def astar(graph, heuristic, start, goal):
    pq = []
    heapq.heappush(pq, (0, start, 0, [start]))
    visited = set()

    while pq:
        f, node, g, path = heapq.heappop(pq)

        if node == goal:
            return path, g

        if node in visited:
            continue
        visited.add(node)

        for neighbor, cost in graph[node]:
            new_g = g + cost
            new_f = new_g + heuristic[neighbor]
            heapq.heappush(pq, (new_f, neighbor, new_g, path + [neighbor]))

    return None, float('inf')

def bfs(graph, start, goal):
    queue = deque([(start, [start], 0)])
    visited = set()

    while queue:
        node, path, cost = queue.popleft()

        if node == goal:
            return path, cost

        if node in visited:
            continue
        visited.add(node)

        for neighbor, edge_cost in graph[node]:
            queue.append((neighbor, path + [neighbor], cost + edge_cost))

    return None, float('inf')

start_time = time.time()
astar_path, astar_cost = astar(graph, heuristic, start, goal)
astar_time = time.time() - start_time

start_time = time.time()
bfs_path, bfs_cost = bfs(graph, start, goal)
bfs_time = time.time() - start_time

print("A* Algorithm:")
print("Path:", " -> ".join(astar_path))
print("Cost:", astar_cost)
print("Time taken:", astar_time)

print("\nBFS Algorithm:")
print("Path:", " -> ".join(bfs_path))
print("Cost:", bfs_cost)
print("Time taken:", bfs_time)

print("\nComparison:")
if astar_time < bfs_time:
    print("A* is faster than BFS")
else:
    print("BFS is faster than A*")

if astar_cost == bfs_cost:
    print("Both found optimal cost")
else:
    print("A* gives better (optimal) cost")
```

```python
#Practical 4
import math

board = [' ' for _ in range(9)]

def print_board():
    for i in range(0, 9, 3):
        print(board[i], '|', board[i+1], '|', board[i+2])
    print()

def check_winner(b, player):
    win_conditions = [
        [0,1,2],[3,4,5],[6,7,8],
        [0,3,6],[1,4,7],[2,5,8],
        [0,4,8],[2,4,6]
    ]
    for condition in win_conditions:
        if b[condition[0]] == b[condition[1]] == b[condition[2]] == player:
            return True
    return False

def is_full(b):
    return ' ' not in b

def minimax(b, depth, is_maximizing):
    if check_winner(b, 'O'):
        return 1
    if check_winner(b, 'X'):
        return -1
    if is_full(b):
        return 0

    if is_maximizing:
        best = -math.inf
        for i in range(9):
            if b[i] == ' ':
                b[i] = 'O'
                score = minimax(b, depth + 1, False)
                b[i] = ' '
                best = max(best, score)
        return best
    else:
        best = math.inf
        for i in range(9):
            if b[i] == ' ':
                b[i] = 'X'
                score = minimax(b, depth + 1, True)
                b[i] = ' '
                best = min(best, score)
        return best

def best_move():
    best_score = -math.inf
    move = -1
    for i in range(9):
        if board[i] == ' ':
            board[i] = 'O'
            score = minimax(board, 0, False)
            board[i] = ' '
            if score > best_score:
                best_score = score
                move = i
    return move

def play():
    print("Positions are 0 to 8")
    print_board()

    while True:
        pos = int(input("Enter your move (X): "))
        if board[pos] != ' ':
            print("Invalid move")
            continue
        board[pos] = 'X'
        print_board()

        if check_winner(board, 'X'):
            print("You win!")
            break
        if is_full(board):
            print("Draw!")
            break

        ai_move = best_move()
        board[ai_move] = 'O'
        print("AI move:")
        print_board()

        if check_winner(board, 'O'):
            print("AI wins!")
            break
        if is_full(board):
            print("Draw!")
            break

play()
```

```python
#Practical 5
import itertools

letters = ('S','E','N','D','M','O','R','Y')
digits = range(10)

for perm in itertools.permutations(digits, len(letters)):
    mapping = dict(zip(letters, perm))

    if mapping['S'] == 0 or mapping['M'] == 0:
        continue

    send = mapping['S']*1000 + mapping['E']*100 + mapping['N']*10 + mapping['D']
    more = mapping['M']*1000 + mapping['O']*100 + mapping['R']*10 + mapping['E']
    money = mapping['M']*10000 + mapping['O']*1000 + mapping['N']*100 + mapping['E']*10 + mapping['Y']

    if send + more == money:
        print("Solution Found:")
        for k in mapping:
            print(k, "=", mapping[k])
        print("SEND =", send)
        print("MORE =", more)
        print("MONEY =", money)
        break
```

```python
#Practical 6
import math

tree = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F', 'G'],
    'D': 3,
    'E': 5,
    'F': 6,
    'G': 9
}

def alphabeta(node, depth, alpha, beta, maximizing):
    if isinstance(tree[node], int):
        return tree[node]

    if maximizing:
        value = -math.inf
        for child in tree[node]:
            value = max(value, alphabeta(child, depth+1, alpha, beta, False))
            alpha = max(alpha, value)
            if beta <= alpha:
                break
        return value
    else:
        value = math.inf
        for child in tree[node]:
            value = min(value, alphabeta(child, depth+1, alpha, beta, True))
            beta = min(beta, value)
            if beta <= alpha:
                break
        return value

result = alphabeta('A', 0, -math.inf, math.inf, True)

print("Optimal value:", result)
```

```python
#Practical 8
knowledge_base = []

def add_fact(predicate, *args):
    knowledge_base.append((predicate, args))

def query(predicate, *args):
    results = []
    for fact in knowledge_base:
        if fact[0] == predicate:
            if all(a == b or b is None for a, b in zip(fact[1], args)):
                results.append(fact)
    return results

add_fact("is_a", "cat", "animal")
add_fact("is_a", "animal", "living_being")
add_fact("has_property", "animal", "breathes")
add_fact("has_property", "cat", "meows")
add_fact("instance_of", "tom", "cat")

def infer_property(entity, prop):
    for fact in knowledge_base:
        if fact == ("has_property", (entity, prop)):
            return True
    for fact in knowledge_base:
        if fact[0] == "instance_of" and fact[1][0] == entity:
            if infer_property(fact[1][1], prop):
                return True
    for fact in knowledge_base:
        if fact[0] == "is_a" and fact[1][0] == entity:
            if infer_property(fact[1][1], prop):
                return True
    return False

print("Facts:")
for f in knowledge_base:
    print(f)

print("\nQuery: What is Tom?")
print(query("instance_of", "tom", None))

print("\nDoes Tom meow?")
print(infer_property("tom", "meows"))

print("\nDoes Tom breathe?")
print(infer_property("tom", "breathes"))
```

```python
#Practical 9
def predict_rain(humidity, temperature, wind, clouds):
    if humidity > 70 and clouds == "high":
        return "Rainy"
    if humidity > 60 and temperature < 25 and clouds == "medium":
        return "Rainy"
    if wind == "strong" and clouds == "high":
        return "Rainy"
    if humidity < 40 and clouds == "low":
        return "No Rain"
    return "No Rain"

humidity = int(input("Enter humidity (%): "))
temperature = int(input("Enter temperature (°C): "))
wind = input("Enter wind (weak/strong): ")
clouds = input("Enter cloud level (low/medium/high): ")

result = predict_rain(humidity, temperature, wind, clouds)

print("Prediction:", result)
```

```python
#Practical 11
import numpy as np

X = np.array([1, 2, 3, 4, 5], dtype=float)
Y = np.array([2, 4, 5, 4, 5], dtype=float)

m = 0
b = 0
lr = 0.01
epochs = 1000
n = len(X)

for _ in range(epochs):
    Y_pred = m * X + b
    dm = (-2/n) * np.sum(X * (Y - Y_pred))
    db = (-2/n) * np.sum(Y - Y_pred)
    m = m - lr * dm
    b = b - lr * db

print("Slope (m):", m)
print("Intercept (b):", b)

x_new = float(input("Enter value to predict: "))
y_new = m * x_new + b

print("Predicted value:", y_new)
```
